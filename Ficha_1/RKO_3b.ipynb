## 3 b) RKO - Reaction Knockout 

```python
#Importação das bibliotecas e leitura do modelo
from cobra.io import read_sbml_model
model = read_sbml_model('iML1515.xml')
envcond = {'EX_glc__D_e': (-15.0, 100000.0),
           'EX_o2_e':(0,1000)}

#Definição do modelo
PRODUCT_ID = 'EX_for_e'
BIOMASS_ID = 'BIOMASS_Ec_iML1515_core_75p37M'

#Importação da biblioteca para a otimização
from mewpy.optimization.evaluation import  BPCY, TargetFlux, WYIELD, BPCY_FVA

#Definir as funções objetivo para a otimização
funcao_1 = BPCY(BIOMASS_ID, PRODUCT_ID, method="pFBA")
funcao_2 = WYIELD(BIOMASS_ID, PRODUCT_ID)

#Definição dos problemas
from mewpy.problems import RKOProblem
problem = RKOProblem(model, fevaluation = [funcao_1, funcao_2], envcond=envcond, candidate_max_size=5)

from mewpy.optimization import EA
ea = EA(problem, max_generations= 20)
resolucao = ea.run()
print(resolucao)

gene_list.close()
#Impressão de uma lista de knockout de reações
gene_list = open('RKO_list.txt', 'w')
for i in resolucao:
    gene_list.write(str(i) + '\n')

gene_list.close()
```
